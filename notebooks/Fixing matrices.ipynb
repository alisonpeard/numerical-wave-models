```python
import numpy as np
import scipy.sparse as sp
import matplotlib.pyplot as plt
import swan

# https://math.mit.edu/~stevenj/18.303/lecture-10.html
# do scipy sparse convesions later

A = swan.upwind(5)
plt.spy(A)
```

```python
nx = 3
ny = 3

cx = np.linspace(0, 2, nx * ny)
cy = np.ones(nx * ny) #np.sin(np.arange(0, nx * ny))

Ax = swan.upwind(nx)
Ay = swan.upwind(ny)

A_spatial = swan.kronecker_product(Ax, Ay, cx, cy)
print(Ax.shape, Ay.shape, A.shape)
fig, axs = plt.subplots()
axs.imshow(A_spatial)
```

# 4D finite difference matrices

```python
mu = 1

ntheta = 4
nsigma = 4

ctheta = np.linspace(-2, 2, ntheta)
csigma = np.ones(nsigma) #np.linspace(0, 5, nsigma)

Atheta = swan.hybrid(ctheta, mu)
Asigma = swan.hybrid(csigma, mu)


fig, axs = plt.subplots(1, 2, figsize=(8, 3))
axs[0].imshow(Atheta)
im = axs[1].imshow(Asigma)
plt.colorbar(im);

fig, ax = plt.subplots()
A_spectral = swan.kronecker_product(Asigma, Atheta, 1, 1)
ax.imshow(A_spectral)
```
