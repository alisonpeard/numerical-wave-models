## 2D wave evolution in $(x,y)$-direction

$$\frac{\partial N}{\partial t} + \frac{\partial_{c_x} N}{\partial x} + \frac{\partial_{c_y} N}{\partial y} = S_\text{tot} \qquad (3.1) $$

```python
import numpy as np
import matplotlib.pyplot as plt
from variables import *
from importlib import reload


def bsbt_matrix(nx, ny, cx, cy, dx, dy):
    """Construct BSBT scheme matrix."""
    A = np.diag((cx.ravel() / dx) + (cy.ravel() / dy))
    A -= np.diag((cx.ravel()[1:] / dx), -1)
    A -= np.diag((cy.ravel()[nx:] / dy), -nx)
    return A


def sordup_matrix(nx, ny, cx, cy, dx, dy):
    """Construct SORDUP scheme matrix."""
    A = np.diag(3 * (cx.ravel() / (2 * dx)) + 3 * (cy.ravel() / (2 * dy)))
    # x updates
    A -= np.diag(4 * (cx.ravel()[1:] / (2 * dx)), -1)
    A += np.diag((cx.ravel()[2:] / (2 * dx)), -2)
    # y updates
    A -= np.diag(4 * (cy.ravel()[nx:] / (2 * dy)), -nx)
    A += np.diag((cy.ravel()[2 * nx:] / (2 * dy)), -2 * nx)
    return A
```

# Spectral solution for single grid cell

```python
def group_velocity(k, d, , u):
    """cg = (1 / 2) * (1 + (2 * l2(k) * d) / np.sinh(2 * l2(k) * d)) * ( * k) / l2(k)**2 + u"""
     = np.array()
    cg = (1 / 2) * (1 + (2 * l2(k) * d) / np.sinh(2 * l2(k) * d))
    cg = cg[..., np.newaxis] * np.einsum('n,md->mnd',,k) / l2(k)
    cg = cg + u
    return cg


def hybrid_matrix_theta(c, step, nu=0.5):
    """nu=0: upwind scheme, nu=1: central difference scheme.
    step should be 1 for row-major ('C') ordering in (,).
    """
    c = c.ravel(order='C').copy() # numpy arrays are mutable
    counter_clockwise = c >= 0
    
    coeff_diag = np.where(counter_clockwise, 1 - 0.5 * nu, 0.5 * nu)
    coeff_super = np.where(counter_clockwise, 0.5 * nu, 1 - 0.5 * nu)
    left = coeff_diag * np.diag(c)
    left += coeff_super * np.diag(c[:-step], step)
    
    coeff_diag = np.where(counter_clockwise, 0.5 * nu, 1 - 0.5 * nu)
    coeff_sub = np.where(counter_clockwise, 1 - 0.5 * nu, 0.5 * nu)
    right = coeff_diag * np.diag(c[step:], -step)
    right += coeff_sub * np.diag(c)
    
    A = left - right
    return A


def hybrid_matrix_sigma(c, step, mu=0.5):
    """mu=0: upwind scheme, nu=1: central difference scheme.
    step should be 1 for row-major ('C') ordering in (,).
    """
    c = c.ravel(order='C').copy() # numpy arrays are mutable
    higher = c >= 0
    
    coeff_diag = np.where(higher, 1 - 0.5 * mu, 0.5 * mu)
    coeff_super = np.where(higher, 0.5 * mu, 1 - 0.5 * mu)
    left = coeff_diag * np.diag(c)
    left += coeff_super * np.diag(c[:-step], step)
    
    coeff_diag = np.where(higher, 0.5 * mu, 1 - 0.5 * mu)
    coeff_sub = np.where(higher, 1 - 0.5 * mu, 0.5 * mu)
    right = coeff_diag * np.diag(c[step:], -step)
    right += coeff_diag * np.diag(c)
    
    A = left - right
    return A
```

```python
# change sign of c to refract clockwise(-)/counterclockwise(+)
c = -np.ones([n, n])
c = np.random.uniform(-1, 1, [n, n])

N = np.zeros([n, n])

ntheta = int(n / 4)
sweeps = [sweep1, sweep2, sweep3, sweep4]
flips = [lambda x: x, lambda x: np.flip(x, 1), lambda x: np.flip(x, 0), lambda x: np.flip(x, None)]

N1 = N.copy() #to store results
for sweep, flip in zip(sweeps, flips):
    Nj = N[sweep, :] #extract indices
    Nj[2, 2] = 1
    Nj = flip(Nj).ravel()   # rotate accordingly
    
    # calculate update matrix
    A = hybrid_matrix_theta(c[sweep, :], step=n) + hybrid_matrix_sigma(c[sweep, :], step=1)
    I = np.eye(n * ntheta)
    Ainv = np.linalg.inv(I + dt * A)

    Nj = Ainv @ Nj         # implicit Euler in time
    #Nj -= dt * (A @ Nj)   # explicit Euler in time
    N1[sweep, :] = flip(Nj.reshape([ntheta, n]))
    
fig, axs = plt.subplots(1, 1)
im = axs.imshow(N1, cmap='YlGnBu')
axs.set_xlabel(r'$\sigma$')
axs.set_ylabel(r'$\theta$')
axs.set_yticks(range(n), labels=labels)
plt.colorbar(im)
```
